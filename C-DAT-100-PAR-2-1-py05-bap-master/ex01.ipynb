```python
import pandas as pd
import json
```

```python
france = pd.read_csv("2020_french_house_pricing.csv")
```

```python
def get_departments():
    dpt = france["Code departement"]
    df = pd.DataFrame(dpt)
    df_dpt = pd.DataFrame(df["Code departement"].unique())
    return df_dpt.to_json()

get_departments()
```

```python
def get_towns():
    towns = france["Commune"]
    df = pd.DataFrame(towns)
    df_towns = pd.DataFrame(df["Commune"].unique())
    return df_towns.to_json()

get_towns()
```

```python
def get_towns_by_dep(department_id):
    result = france[france["Code departement"] == department_id]                    
    res = pd.DataFrame(result["Commune"].unique())
    return res #.to_json()

get_towns_by_dep(13)
```

```python
def get_prices(department_id, town):
    result = france[(france["Code departement"] == department_id) 
                    & (france["Commune"] == town) 
                    & (france["Nature mutation"] == "Vente")]
    res = pd.DataFrame(result["Valeur fonciere"])
    return res.to_json()

get_prices(13, "MARSEILLE 8EME")
```

```python
def get_price(department_id, town, rooms_min, rooms_max):
    result = france[(france["Code departement"] == department_id) 
                    & (france["Commune"] == town) 
                    & (france["Nature mutation"] == "Vente") 
                    & (france["Nombre pieces principales"].between(rooms_min, rooms_max))]
    res = pd.DataFrame(result["Valeur fonciere"])
    return res.to_json()

get_price(13, "MARSEILLE 8EME", 3, 4)
```

```python
def get_price(department_id, town, rooms_min, rooms_max, sq_min, sq_max):
    result = france[(france["Code departement"] == department_id) 
                    & (france["Commune"] == town) 
                    & (france["Nature mutation"] == "Vente") 
                    & (france["Nombre pieces principales"].between(rooms_min, rooms_max)) 
                    & (france["Surface reelle bati"].between(sq_min, sq_max))]
    res = pd.DataFrame(result["Valeur fonciere"])
    return res.to_json()

get_price(13, "MARSEILLE 8EME", 3, 4, 60, 70)
```
