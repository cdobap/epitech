EXERCICE 1
###done with ebay -> webscrapings'problems with leboncoin###
random days to get dates like leboncoin

```python
url = "https://www.ebay.fr/sch/i.html?_from=R40&_nkw=ps5&_sacat=139971&_ipg=25"
response = rq.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
products = soup.findAll('div', class_='s-item__info clearfix')

titles = []
details = []
prices = []
days = []
day_list= ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

def get_ps5_prices():      
    for product in products:     
        title = product.find('h3').get_text()
        detail = product.find('span', class_='SECONDARY_INFO')
        price = product.find('span', class_='s-item__price')
        if len(title) > 1 and len(str(price)[28:-11:]) < 7:
            titles.append(title)
            details.append(str(detail)[29:-7:])
            prices.append(int(str(price)[28:-14:]))
            days.append(random.choice(day_list))
    
get_ps5_prices()

df = pd.DataFrame(
    {
        "title": titles,
        "detail": details,
        "price": prices,
        "day": days,
    }
)

df
```

LOOP TO CHECK CHANGE
due to random date, there is always changes,
comment "day" here and inside the function to get the same dataframe 

```python
while True:
    titles = []
    details = []
    prices = []
    days = []
    get_ps5_prices()

    df1 = pd.DataFrame(
        {
            "title": titles,
            "detail": details,
            "price": prices,
            "day": days,
        }
    )
    if df1.equals(df):
        print("no change")
    else:
        file = open('ps5-dataframe.pickle', 'wb')
        pickle.dump(df1, file)
        file.close()
        print("upload")
    time.sleep(300)
```
