```python
import pandas as pd
import nltk
import string
import re
```

```python
df = pd.read_csv('./csv/twitter/transform/merge_tweets_es_previousdaytrend.csv')
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
counter = df.count(axis=1)
```

```python
counter
```

```python
group_tweets = []

for row in df.index:
    group_tweets.append(' '.join(str(x) for x in df.iloc[row, 7:1387]))
```

```python
group_tweets[4]
```

```python
clean_tweets = []

for i in range (0, len(group_tweets)):
    tw = group_tweets[i]
    t = re.sub('nan', '', tw)
    t = re.sub('\n', '', t)
    t = re.sub('@', '', t)
    t = re.sub('#', '', t)
    t = re.sub(r'http\S+', '', t)
    t = t.lower()    
    " ".join(t.split())
    translator = str.maketrans('', '', string.punctuation)
    t.translate(translator)
    clean_tweets.append(t)
    
```

```python
clean_tweets[4]
```

```python
df["clean_text"] = clean_tweets
df
```

```python
filename = './csv/twitter/result/clean_tweets_es_trends.csv'
df.to_csv(filename)
```
