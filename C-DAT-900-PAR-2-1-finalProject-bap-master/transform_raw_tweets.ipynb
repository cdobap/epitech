```python
df5 = pd.read_csv('./csv/twitter/transform/transform_tweets_onlydate.csv')
df5 = df5[['onlydate', 'text']]
df5
```

```python
data = {}

for i in df5.index:
    if df5['onlydate'][i] in data:
        #print('exist')
        #print(df5['onlydate'][i])       
        data[df5['onlydate'][i]].append(df5['text'][i])        
    else:
        #print('no')
        data[df5['onlydate'][i]] = [df5['text'][i]]
    
#data
    
```

```python
df6 = pd.DataFrame.from_dict(data, orient='index')
df6
```

```python
filename = './csv/twitter/transform/ready_tweets_onlyonedate_per_row.csv'
df6.to_csv(filename)
```
